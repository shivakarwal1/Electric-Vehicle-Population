### What is highest electric range by car?
---



```python
#Find the highest range value
max_range = df['Electric Range'].max()

#Find the vehicle with the highest range
vehicle_with_max_range = df[df['Electric Range'] == max_range]['Make'].values[0]

#Plot all ranges
plt.figure(figsize=(12, 6))
plt.bar(df['Make'], df['Electric Range'], color='grey')  # Use a neutral color like grey for all other bars

#Highlight the bar for the vehicle with the highest range
plt.bar(vehicle_with_max_range, max_range, color='blue')

#Adding labels and title
plt.title('Electric Range by Vehicle')
plt.xlabel('Vehicle Make')
plt.ylabel('Electric Range (miles)')
plt.xticks(rotation=90)  # Rotate the x-axis labels for better readability if needed

plt.show()
```

# Can we observe a trend in the charging duration for electric vehicles based on the day of the week, possibly due to different usage patterns?

```python
import matplotlib.pyplot as plt
import pandas as pd

daily_charging_durations = [2.5, 3, 2.8, 3.1, 2.7, 4, 3.5]  # Hours charged each day

days_of_the_week = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

plt.figure(figsize=(10, 6))
plt.plot(days_of_the_week, daily_charging_durations, marker='o', linestyle='-', color='blue')
plt.title('Charging Duration for Electric Vehicles During the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Charging Duration (Hours)')
plt.grid(True)
plt.show()
```

# Which county has the most electric cars over time?

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path = '/content/Electric_Vehicle_Population_Data.csv' # Change this to your file path
data = pd.read_csv(file_path)

county_year_count = data.groupby(['County', 'Model Year']).size().reset_index(name='Count')

pivot_data = county_year_count.pivot(index='Model Year', columns='County', values='Count').fillna(0)

plt.figure(figsize=(15, 8))
pivot_data.plot(kind='bar', stacked=True)
plt.title('Number of Electric Cars by County Over Time')
plt.ylabel('Number of Electric Cars')
plt.xlabel('Model Year')
plt.legend(title='County', bbox_to_anchor=(0.5, -0.15), loc='upper center', ncol=4)
plt.tight_layout()
plt.show()
```

#How many miles can a car go on one single charge?

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(12, 6))
plt.hist(data['Electric Range'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Electric Range (Miles on a Single Charge)')
plt.xlabel('Electric Range (Miles)')
plt.ylabel('Number of Cars')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

#Which cars best fit the need of a customer?
